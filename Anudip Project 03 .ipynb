```python
# Exploring Restaurant Types
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

```python
# Online Order Availability
```

```python
sns.countplot(x=dataframe['online_order'])
```

```python
# Identify the Most Popular Restaurant
```

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']

print('Restaurant(s) with the maximum votes:')
print(restaurant_with_max_votes)
```

```python
#Votes by Restaurant Type
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel('Type of restaurant', c='red', size=20)
plt.ylabel('Votes', c='red', size=20)
sns.countplot(x=dataframe['online_order'])
```

```python
#Analyzing food type
```

```python
plt.hist(dataframe['food_type'], bins=5)
plt.title('Bookings based on food type')
plt.show()
```

```python
#Approximate cost for two people
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
# Food type vs Ratings
```

```python
plt.figure(figsize=(6,5))
plt.scatter(dataframe['food_type'], dataframe['rate'], color='red')
plt.title("Food type vs Rating")
plt.xlabel("Food tpye")
plt.ylabel("Rating")
plt.show()
```
